```python
def solve():
    import os

    if not os.path.exists("input.txt"):
        print("input.txt not found")
        return

    # Read and strip newlines, keep spacing
    with open("input.txt", "r") as f:
        lines = [line.rstrip("\n") for line in f]

    # Last line has operators
    op_line = lines[-1]
    num_lines = lines[:-1]

    # Pad all lines to equal width
    max_len = max(len(line) for line in lines)
    num_lines = [line.ljust(max_len) for line in num_lines]
    op_line = op_line.ljust(max_len)

    # Identify separator columns â€” full column of spaces
    separator_cols = set()

    for col in range(max_len):
        if all(row[col] == " " for row in num_lines):  # Only check number rows
            separator_cols.add(col)

    problems = []
    col = 0

    # Group into problem-column segments
    while col < max_len:
        if col in separator_cols:
            col += 1
            continue

        start = col
        while col < max_len and col not in separator_cols:
            col += 1
        end = col

        # Extract operator
        op = op_line[start:end].strip()
        if op not in "+*":
            continue

        # Extract numbers (one per row)
        numbers = []
        for row in num_lines:
            text = row[start:end].strip()
            if text.isdigit():
                numbers.append(int(text))

        if numbers:
            problems.append((numbers, op))

    # Compute grand total
    grand_total = 0

    for numbers, op in problems:
        if op == "+":
            grand_total += sum(numbers)
        else:
            product = 1
            for n in numbers:
                product *= n
            grand_total += product

    print("Part 1 - Number of problems:", len(problems))
    print("Part 1 - Grand Total:", grand_total)

    return grand_total
```

```python
solve()
```

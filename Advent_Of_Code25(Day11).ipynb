```python
from collections import defaultdict

# -------- READ input.txt --------
with open("input.txt") as f:
    lines = [line.strip() for line in f if line.strip()]

# -------- PARSE GRAPH --------
graph = defaultdict(list)

for line in lines:
    left, right = line.split(":")
    device = left.strip()
    outputs = right.strip().split()
    graph[device] = outputs

# -------- FIND NUMBER OF PATHS FROM 'you' TO 'out' --------
start = "you"
end = "out"

count = 0

def dfs(node):
    global count
    if node == end:
        count += 1
        return
    for nxt in graph[node]:
        dfs(nxt)

dfs(start)

print(count)
```

```python
from collections import defaultdict

# ---------- Build graph ----------
graph = defaultdict(list)
with open("input.txt") as f:
    for line in f:
        line = line.strip()
        if not line or ":" not in line:
            continue
        node, neighbors = line.split(":", 1)
        node = node.strip()
        neighbors = neighbors.strip().split()
        graph[node].extend(neighbors)

print("Graph loaded. Nodes:", len(graph))

# ---------- DP memoization ----------
start = "svr"
end = "out"
required_nodes = {"dac", "fft"}

memo = {}

def count_paths(node, visited_required):
    key = (node, frozenset(visited_required))
    if key in memo:
        return memo[key]

    # update visited_required
    new_visited = visited_required | ({node} & required_nodes)

    if node == end:
        if new_visited == required_nodes:
            return 1
        else:
            return 0

    total = 0
    for neighbor in graph.get(node, []):
        total += count_paths(neighbor, new_visited)

    memo[key] = total
    return total

total_paths = count_paths(start, set())
print("Total paths from svr to out visiting BOTH dac and fft:", total_paths)
```

```python
def parse_input(text):
    lines = [l.rstrip() for l in text.splitlines()]
    shapes = {}
    regions = []
    i = 0

    while i < len(lines):
        line = lines[i]

        # shape start
        if line.endswith(":") and line[:-1].isdigit():
            idx = int(line[:-1])
            i += 1
            block = []
            while i < len(lines) and lines[i] and not ("x" in lines[i] and ":" in lines[i]):
                block.append(lines[i])
                i += 1
            shapes[idx] = block

        # region line
        elif "x" in line and ":" in line:
            dims, cnt = line.split(":")
            w, h = map(int, dims.split("x"))
            counts = list(map(int, cnt.split()))
            regions.append((w, h, counts))
            i += 1

        else:
            i += 1

    return shapes, regions


def shape_area(shape):
    return sum(row.count("#") for row in shape)


def shape_min_bbox(shape):
    # find bounding box dimensions
    rows = shape
    h = len(rows)
    w = len(rows[0])
    cells = [(r,c) for r in range(h) for c in range(w) if rows[r][c] == "#"]
    if not cells:
        return (0,0)
    min_r = min(r for r,c in cells)
    max_r = max(r for r,c in cells)
    min_c = min(c for r,c in cells)
    max_c = max(c for r,c in cells)
    return (max_r - min_r + 1, max_c - min_c + 1)  # height, width


def can_fit_region(w, h, shapes, counts):
    total_area = 0

    for idx, cnt in enumerate(counts):
        if cnt == 0:
            continue

        area = shape_area(shapes[idx])
        total_area += area * cnt

        # check bounding box must fit (either orientation)
        sh, sw = shape_min_bbox(shapes[idx])
        if not ((sw <= w and sh <= h) or (sh <= w and sw <= h)):
            return False  # shape cannot fit at all

    return total_area <= w * h


# MAIN
with open("input.txt") as f:
    text = f.read()

shapes, regions = parse_input(text)

answer = sum(
    1 for (w, h, counts) in regions
    if can_fit_region(w, h, shapes, counts)
)

print("Part 1:", answer)
```
